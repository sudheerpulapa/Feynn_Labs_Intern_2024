```python
# Importing necessary libraries
import pandas as pd  # For data manipulation and analysis
import matplotlib.pyplot as plt  # For creating visualizations
import seaborn as sns  # For statistical data visualization
import numpy as np  # For numerical operations

from sklearn.model_selection import train_test_split  # For splitting data into training and testing sets
from sklearn.preprocessing import StandardScaler  # For standardizing features
from sklearn.decomposition import PCA  # For Principal Component Analysis
from sklearn.cluster import KMeans  # For KMeans clustering algorithm

import warnings  # For suppressing warnings
warnings.filterwarnings("ignore")  # Ignore any warnings
```

```python
# Load EV Charging Station Dataset from CSV file
df2 = pd.read_excel("D:\\Project-2.2\\Datasets\\2.EV-Charging-Station-Dataset.xlsx",
                   sheet_name = "Table 4",
                   header = 1)
# Display the first few rows of the dataset
df2.head()
```

## **Available Retail Outlets for Charging Electric Vehicles in India:**

```python
# Set the style for seaborn
sns.set_style("whitegrid")

# Plotting the point plot
plt.figure(figsize=(10, 8))
sns.pointplot(data=df2.sort_values(by='EV Charging Facility', ascending=False), y='State/UT', x='EV Charging Facility', color='orange')
plt.xlabel('Number of Retail Outlets for Charging EVs', fontsize=14, family='serif', labelpad=10)
plt.ylabel('State/UT', fontsize=14, family='serif')
plt.title('Available Retail Outlets for Charging EVs in India', fontsize=15, family='serif', weight='bold', pad=12)
plt.xticks(fontfamily='serif', size=12)
plt.yticks(fontfamily='serif', size=12)
plt.tick_params(direction='inout')
plt.tight_layout()

# Save the figure
plt.savefig('retail_outlets_visualization.png', dpi=300)

# Show the plot
plt.show()
```
